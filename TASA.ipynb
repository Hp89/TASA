```python
from bs4 import BeautifulSoup
import urllib.request
import re
def scrap():
    page = urllib.request.urlopen(r'https://timesofindia.indiatimes.com/india/reactions-to-bjp-win-expose-cpm-rift/articleshow/62127134.cms')
    data = page.read()
    soup = BeautifulSoup(data,'html.parser')
    pst = str(soup.find_all("p"))
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr,'',pst)
    textwithsomeerror = cleantext[0:len(cleantext)-1]
    print(textwithsomeerror)
    print(True)
```
